## Решение задания 3

````python
months_names = ['январь', 'февраль', 'март', 'апрель', 'май', 'июнь', 'июль', 'август', 'сентябрь', 'октябрь', 'ноябрь', 'декабрь']
months_numbers = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
error = False
error_text = ''

user_day = input('Введите день: ')
# Если день с 1 по 9, дополняем его слева нулем
# Проверяем, что введено число от 1 до 31
if 1 <= int(user_day) <= 31 :
    if len(user_day)==1 :
        user_day = '0' + user_day    
else:
    error_text = 'Введенное число содержит ошибку'
    error = True
    
user_month = input('Введите название месяца: ')

# Проверяем, что введенное пользователем название месяца корректно
if user_month.lower() in months_names :
    month_number = months_numbers[months_names.index(user_month.lower())]
else :
    error_text = error_text + ' Введенное название месяца ' + user_month + ' содержит ошибку'
    error = True

# Проверку корректности даты нужно делать для каждого месяца отдельно, поскольку в каждом месяце разное число дней
# В этом коде такой проверки нет
    
if error :
    print(error_text)
else :
    # Получаем дату в формате ммдд 
    user_date = month_number + user_day  

    # Реализована конструкция из серии if elif else. Можно было бы использовать вместо этого цикл, перебирающий элементы массива
    if '0320' < user_date < '0420' :
        zodiak_sign = 'Овен'
    elif '0419' < user_date < '0521' :
        zodiak_sign = 'Телец'
    elif '0520' < user_date < '0621' :
        zodiak_sign = 'Близнецы'
    elif '0620' < user_date < '0723' :
        zodiak_sign = 'Рак'
    elif '0722' < user_date < '0823' :
        zodiak_sign = 'Лев'
    elif '0822' < user_date < '0923' :
        zodiak_sign = 'Дева'
    elif '0922' < user_date < '1023' :
        zodiak_sign = 'Весы'
    elif '1022' < user_date < '1122' :
        zodiak_sign = 'Скорпион'
    elif '1121' < user_date < '1222' :
        zodiak_sign = 'Стрелец'
    elif '1221' < user_date < '0121' :
        zodiak_sign = 'Козерог'
    elif '0120' < user_date < '0219' :
        zodiak_sign = 'Водолей'
    elif '0218' < user_date < '0321' :
        zodiak_sign = 'Рыбы'
    else :    
        zodiak_sign = 'Ошибка в введенных данных'

    print('Ваш знак зодика: ', zodiak_sign)    
````

- если длина товара больше 2 метров, то выводите "Упаковка для лыж";


## Решение задания 4

````python
# Примеры исходных данных

# width = 15
# length = 50
# height = 15

# width = 45
# length = 205
# height = 45

# width = 5
# length = 12
# height = 10

# width = 5
# length = 25
# height = 90

width = 300
length = 12
height = 100

# Второе и третье условия задачи могут выполняться одновременно, 
# поэтому я принял решение более важным считать третье условие более приоритетным

# Проверка первого условия: все измерения меньше или равны 15 см
if width <= 15 and length <= 15 and height <= 15:
    print('Коробка №1')
# Проверка третьего условия: длина больше 2 метров (200 см)    
elif length > 200:
    print('Упаковка для лыж')
# Проверка второго условия: хотя бы одно из измерений больше 15 и меньше 50 см
elif (15 < width < 50) or (15 < length < 50) or (15 < height < 50):
    print('Коробка №2')
# Если не выполняется ни одно из предыдущих трех условий, идем сюда
else:
    print('Стандартная коробка №3')

    
````

## Задание 5 (необязательное)

Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет "счастливым". Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.   

Примеры работы программы:  
1. ```number = 123456```

Результат:  
`Несчастливый билет`
2. ```number = 123321```

Результат:  
`Счастливый билет`


___
## Решение задачи 5

````python
# Поскольку в первом блоке по Python рассматривались только примеры математических операций и условия,
# приведенное ниже решение построено на использовании этих возможностей языка

ticket_num = int(input('Введите шесть цифр - номер вашего билета: '))

digit_1 = ticket_num // 100000
ticket_num_part_1 = ticket_num - digit_1 * 100000

digit_2 = ticket_num_part_1 // 10000
ticket_num_part_2 = ticket_num_part_1 - digit_2 * 10000

digit_3 = ticket_num_part_2 // 1000
ticket_num_part_3 = ticket_num_part_2 - digit_3 * 1000

digit_4 = ticket_num_part_3 // 100
ticket_num_part_4 = ticket_num_part_3 - digit_4 * 100

digit_5 = ticket_num_part_4 // 10
digit_6 = ticket_num_part_4 - digit_5 * 10

# print(digit_1, digit_2, digit_3, digit_4, digit_5, digit_6)

if (digit_1 + digit_2 + digit_3) == (digit_4 + digit_5 + digit_6):
    print('Счастливый билет')
else:    
    print('Несчастливый билет')
````

````python
# Альтернативный вариант :)
ticket_num = input('Введите шесть цифр - номер вашего билета: ')

if (int(ticket_num[0]) + int(ticket_num[1]) + int(ticket_num[2])) == (int(ticket_num[3]) + int(ticket_num[4]) + int(ticket_num[5])):
    print('Счастливый билет')
else:    
    print('Несчастливый билет')
````

## Задание 6 (необязательное)

Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник).
Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

- если пользователь выбрал круг, запрашиваем его радиус,
- если треугольник – длины трех его сторон;
- если прямоугольник – длины двух его сторон.

Пример работы программы:
1.  

```
Введите тип фигуры:
Круг

Введите радиус круга:
10
```
Результат:  
`Площадь круга: 314.16`

2.  

```
Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3
```
Результат:  
`Площадь треугольника: 1.98`

___
## Решение задачи 6

````python
figure_type = int(input('Введите цифру для выбора типа фигуры.\n 1 - треугольник, 2 - прямоугольник, 3 - круг: '))

if figure_type == 1:
    # Треугольник
    tr_side_1 = int(input('Длина 1-й стороны: '))
    tr_side_2 = int(input('Длина 2-й стороны: '))
    tr_side_3 = int(input('Длина 3-й стороны: '))
    if tr_side_1 > 0 and tr_side_2 > 0 and tr_side_3 > 0:
        # Считаем площадь треугольника по формуле Герона 
        # https://ru.wikipedia.org/wiki/%D0%A4%D0%BE%D1%80%D0%BC%D1%83%D0%BB%D0%B0_%D0%93%D0%B5%D1%80%D0%BE%D0%BD%D0%B0
        half_perim = (tr_side_1 + tr_side_2 + tr_side_3) / 2
        print('Площать треугольника = ', round((half_perim 
                                                * (half_perim - tr_side_1) 
                                                * (half_perim - tr_side_2) 
                                                * (half_perim - tr_side_3)) 
                                               ** (0.5), 2))
    else:
        print('Значения сторон треугольника введены с ошибками')
elif figure_type == 2:        
    # Прямоугольник        
    rect_side_1 = int(input('Длина 1-й стороны: '))
    rect_side_2 = int(input('Длина 2-й стороны: '))
    print('Площадь прямоугольника = ', str(rect_side_1 * rect_side_2))
elif figure_type == 3:        
    # Круг        
    circle_radius = int(input('Радиус: '))
    print('Площадь круга = ', str(round((3.141572 * circle_radius * circle_radius), 2)))
    
else:        
    # Ошибка
    print('Для выбора фигуры нужно было ввести цифру 1, 2 или 3')
       
````
