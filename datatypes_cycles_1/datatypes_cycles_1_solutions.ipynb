## Задание 2
Напишите программу, которая последовательно запрашивает у пользователя числа (по одному за раз) и после первого нуля выводит сумму всех ранее введенных чисел.

Примеры работы программы:  
1.   

```
Введите число:  
1

Введите число:  
4

Введите число:  
6

Введите число:  
0
```

Результат:  
`11`

2.      

```
Введите число:  
0
```  

Результат:  
`0`

___
## Решение задания 2

````python
# Вначале создаем переменную, в которой будет происходить накопление суммы
user_input_sum = 0
continue_loop = True
# Далее в цикле считываем пользовательское число
print('Эта программа позволяет подсчитать сумму последовательно вводимых чисел. Для окончания подсчета введите 0')
while continue_loop:
    user_input = int(input('Введите число: '))
    if user_input == 0:
        continue_loop = False
    else:
        user_input_sum += user_input
print('Сумма введенных чисел: ', user_input_sum)
````

## Задание 4

У нас есть список, содержащий информацию о среднедневной температуре в Фаренгейтах за произвольный период по странам (структура данных в примере).
Необходимо написать код, который рассчитает среднюю температуру за период в Цельсиях(!) для каждой страны.

Пример работы программы:  
```
countries_temperature = [
    ['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Germany', [57.2, 55.4, 59, 59, 53.6]],
    ['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]
```

Результат:  
```
Средняя температура в странах:
Thailand  -  23.9 С
Germany  -  13.8 С
Russia  -  3.7 С
Poland  -  12.0 С
```

___
## Решение задания 4

````python
# 
countries_temperature = [
    ['Thailand', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Germany', [57.2, 55.4, 59, 59, 53.6]],
    ['Russia', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Poland', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]

if len(countries_temperature) > 0:
    for country_temperatures in countries_temperature:
        country_name = country_temperatures[0]
        # Кажется логичным вначале посчитать среднее значение температуры по Фаренгейту
        country_avg_fahrenheit = sum(country_temperatures[1])/len(country_temperatures[1])
        # а уже потом перевести среднее значение температуры по Фаренгейту в градусы Цельсия
        country_avg_celcius = round((country_avg_fahrenheit - 32) * 5 / 9, 1)
        print(country_name, ' - ', country_avg_celcius, 'C')
else:
    print('Входной массив пуст')
````

## Задание 5 (необязательное)

Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей.

Примеры работы программы:

```
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',

    '2018-03-22,user4,4',

Результат:

```
stream = [
    '2018-01-01,user100,150',

## Решение задания 5

````python
# Задача решена с возможной неоправданной избыточностью, поскольку вначале было не очень внимательно прочитано задание.
# В этом решении считается не среднее количество просмотров на уникального пользователя,
# а среднее количество средних количеств просмотров пользователей, 
# т.е. вначале считается среднее количество для каждого пользователя, а потом - среднее значение этих средних значений.
# Уточненное оптимизированное решение приведено в следующем блоке. Это же стирать стало жалко :)
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11',
]

"""stream = [
    '2018-01-01,user100,150',
    '2018-01-07,user99,205',
    '2018-03-29,user1001,81'
]"""

# Вначале, если лог не пустой, создаем пустые массивы, куда будем складывать информацию о пользователях и их посещениях
if len(stream) > 0:
    users = []
    users_visits = []
    # Первый цикл позволяет перебрать лог и собрать два массива с равным количеством элементов.
    # В первом массиве users будут собраны идентификаторы пользователей
    # Во втором массиве будут собраны массивы с количеством просмотров для каждого из посещений соответствующего пользователя
    for current_record in stream:
        cur_date, cur_user, cur_views = current_record.split(',')
        # Если пользователь ранее уже встречался в логе, мы для него уже успели создать элементы в массивах
        if cur_user in users:
            cur_user_index = users.index(cur_user)
            users_visits[cur_user_index].append(int(cur_views))
        # Если пользователь впервые встречается в логе, создаем для него элементы в массивах users и users_visits    
        else:
            users.append(cur_user)
            users_visits.append([int(cur_views)])
    # Второй цикл позволит перебрать элементы массива users_visits и посчитать суммы посещений по каждому из уникальных
    users_count = len(users)
    loop_counter = 0
    sum_of_average_values = 0

    while loop_counter < users_count:
        sum_of_average_values += sum(users_visits[loop_counter]) / len(users_visits[loop_counter])
        loop_counter += 1
    print('Среднее количество просмотров на уникального пользователя: ', round(sum_of_average_values / users_count, 2))    

else:
    print('Лог посещений пустой')
````

## Задание 6 (необязательное)

Необходимо у пользователя запрашивать набор чисел разделенных пробелом. В результате в отсортированном порядке должны выводиться числа, которые повторяются в вводе более одного раза.

Примеры работы программы:  
1.  

```
Введите числа:
4 8 0 3 4 2 0 3
```

Результат:  
`0 3 4`

2.     

```
Введите числа:
1 1 2 2 3 3
```

Результат:  
`1 2 3`

3.     

```
Введите числа:
10 15 15 103 200 200 200 1 1 1 1 1 2 2 2
```

Результат:  
`1 2 15 200`

___
## Решение задания 6

````python
# Для проверки можно использовать тестовые данные вместо ввода с помощью функции input
# test_sequence = '4 8 0 3 4 2 0 3'
# test_sequence = '1 1 2 2 3 3'
# test_sequence = '10 15 15 103 200 200 200 1 1 1 1 1 2 2 2'

test_sequence = input('Введите последовательность чисел, разделенных пробелом (например: 1 3 4 2 3): ')
# Упаковываем введенную строку в массив
user_numbers = test_sequence.split(' ')

# Создаем два массива. В первый попадают все уникальные номера из введенных пользователем, 
# во второй - те, которые встречаются более одного раза
unique_numbers = []
multiple_numbers = []
# Перебираем массив введенных пользователем чисел
for cur_number in user_numbers:
    if int(cur_number) in unique_numbers:
        # Если число уже встречалось, добавляем его в массив неуникальных чисел, если ранее мы его туда уже не успели добавить
        if int(cur_number) not in multiple_numbers:             
            multiple_numbers.append(int(cur_number))            
    else: # Если число появляется впервые, добавляем его в массив unique_numbers
        unique_numbers.append(int(cur_number))
# Сортируем массив неуникальных чисел (в массиве - int, не str)        
sorted_int_numbers = sorted(multiple_numbers)
# Получаем версию сортированного массива, в котором все элементы - это строки
sorted_str_numbers = [str(multiple_number) for multiple_number in sorted_int_numbers]        
print(' '.join(sorted_str_numbers)) 
````
